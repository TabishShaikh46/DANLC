# ***Diabetes Insights***

```python
# Set the figure size for the plot
plt.figure(figsize=(15, 6))
# Define a color palette for the bars
colors = ["#FF6666", "#FF4D4D", "#FF3333", "#FF1A1A", "#FF8080", "#FF9999", "#FFB3B3"]
# Create a count plot showing the distribution of individuals by region and gender
sns.countplot(data=df, x='Age', hue='Gender', palette=colors)
# Set the title of the plot
plt.title("Age by Gender")
# Label the x-axis
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Count")
# Rotate the x-axis labels for improved readability
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Count the number of patients in each region
region_counts = df['Family_History_Diabetes'].value_counts()
# Set the size of the figure for the pie chart
plt.figure(figsize=(10, 6))
# Define a color palette for the pie chart segments
colors = ["#FF6666", "#FFB3B3"]
# Create a pie chart to visualize the distribution of patients by region
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
# Set the title for the pie chart
plt.title('Diabetes Distribution by Family History')
# Ensure the pie chart is circular by setting equal aspect ratio
plt.axis('equal')
# Display the pie chart
plt.show()
```

```python
# Create a pivot table to summarize BMI counts by region and family history
pivot_table = df.pivot_table(index='Region', columns='Diabetes_Type', values='BMI', aggfunc='mean')
# Set the figure size for the heatmap
plt.figure(figsize=(15, 6))
# Create a heatmap to visualize the BMI data across regions and family history
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f')
# Set the title for the heatmap
plt.title("BMI by Region and Parent Diabetes Type")
# Label the x-axis
plt.xlabel("Diabetes Type")
# Label the y-axis
plt.ylabel("Region")
# Display the heatmap
plt.show()
```

```python
# Set the figure size for the histogram to ensure clarity in visualization
plt.figure(figsize=(15, 6))
# Create a histogram to visualize the distribution of cholesterol levels across different age groups
sns.histplot(data=df, x="Age", bins=20, kde=True,color="#FF1A1A")
# Set the title for the histogram
plt.title("Distribution of Cholesterol Levels by Age")
# Label the x-axis
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Cholesterol (mg/dL)")
# Display the histogram
plt.show()
```

```python
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(12, 6))
# Create a line plot to show the average physical activity level for each region
sns.lineplot(data=df, x="Region", y="Physical_Activity_Level", marker="o", color="#FF1A1A")
# Set the title for the line plot to describe the data being visualized
plt.title("Physical Activity Level by Region")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```

```python
# Set the figure size for the donut chart to ensure it is visually appealing
plt.figure(figsize=(12, 6))
# Count the occurrences of each dietary habit in the dataset
diet_counts = df["Dietary_Habits"].value_counts()
# Create a donut chart to visualize the distribution of dietary habits
plt.pie(diet_counts, labels=diet_counts.index, autopct="%1.1f%%", colors=["#FF8080", "#FF9999", "#FFB3B3"],
        wedgeprops={"width": 0.4})
# Set the title for the donut chart to describe the data being visualized
plt.title("Dietary Habits Distribution")
# Display the donut chart
plt.show()
```

```python
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(18, 6))
# Create a line plot to show the relationship between age and heart rate
sns.lineplot(data=df, x="Age", y="Genetic_Risk_Score", marker="o", color="#FF1A1A")
# Set the title for the line plot to describe the data being visualized
plt.title("Genetic Risk Score by Age")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```

```python
# Set the figure size for the bar plot to ensure clarity in visualization
plt.figure(figsize=(10, 6))
colors = [ "#FF8080", "#FF9999", "#FFB3B3"]
# Create a bar plot to visualize alcohol consumption categorized by region
sns.countplot(data=df, x='Alcohol_Consumption', hue='Gender',palette=colors)
# Set the title for the bar plot to describe the data being visualized
plt.title('Alcohol Consumption by Region')
# Display the bar plot
plt.show()
```

```python
# Create a pivot table to summarize BMI data by region and country
pivot_table = df.pivot_table(index='Region', columns='Gender', values='Fasting_Blood_Sugar', aggfunc='mean')
# Set the figure size for the heatmap to enhance visualization
plt.figure(figsize=(15, 6))
# Create a heatmap to visualize the average BMI data across countries and regions
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f')
# Set the title for the heatmap to describe the data being visualized
plt.title("Fasting Blood Sugar by Gender and Region")
# Label the x-axis
plt.xlabel("Gender")
# Label the y-axis
plt.ylabel("Region")
# Display the heatmap
plt.show()
```

```python
# Set the figure size for the histogram to enhance visibility
plt.figure(figsize=(12, 6))
["#FF6666", "#FF4D4D", "#FF3333", "#FF1A1A", "#FF8080", "#FF9999", "#FFB3B3"]
# Create a histogram to visualize the age distribution categorized by smoking status
sns.histplot(data=df, x='Age', hue='Smoking', multiple='stack', bins=10,palette=colors)
# Set the title for the histogram to describe the data being visualized
plt.title('Age Distribution by Smoking Status')
# Display the histogram
plt.show()
```

```python
# Assuming df is your DataFrame
# Set the figure size for the donut chart to ensure it is visually appealing
plt.figure(figsize=(8, 4))
# Count the occurrences of each HbA1c status in the dataset
# Define the categories based on HbA1c levels
def categorize_hba1c(hba1c):
    if hba1c < 5.7:
        return 'Normal'
    elif 5.7 <= hba1c < 6.5:
        return 'Prediabetes'
    else:
        return 'Diabetes'

# Apply the categorization function to the HbA1c column
hba1c_counts = df['HbA1c'].apply(categorize_hba1c).value_counts()
# Create a donut chart to visualize the distribution of HbA1c status
plt.pie(hba1c_counts, labels=hba1c_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3),
        colors=["#FF6666", "#FF8080", "#FFB3B3"])
# Set the title for the donut chart to describe the data being visualized
plt.title('HbA1c Status Distribution')
# Display the donut chart
plt.show()
```

```python
# Set the figure size for the count plot
plt.figure(figsize=(18, 6))
# Create a count plot to visualize the distribution of Fast Food Intake
sns.countplot(data=df, x='Fast_Food_Intake', palette='Reds',  hue='Fast_Food_Intake',legend=False)
# Set the title for the count plot
plt.title('Count of Fast Food Intake Levels')
# Label the x-axis
plt.xlabel('Fast Food Intake')
# Label the y-axis
plt.ylabel('Count')
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the count plot
plt.show()
```

```python
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(15, 6))
# Group the data by region and calculate the mean stress level for each region
region_stress = df.groupby('Region')['Stress_Level'].mean().reset_index()
# Create a line plot to visualize the average stress level for each region
sns.lineplot(data=region_stress, x="Region", y="Stress_Level", marker="o", color="#FF6666")
# Set the title for the line plot
plt.title("Stress Level by Region")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```
